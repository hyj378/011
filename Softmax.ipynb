```
import numpy as np
xy = np.loadtxt('data-04-zoo.csv', delimiter=',', dtype=np.float32)

x_data = xy[:, :-1]
y_data = xy[:, [-1]]

nb_classes = 7 # 0 ~ 6

y_one_hot = tf.one_hot(y_data, nb_classes)
y_one_hot = tf.reshape(y_one_hot, [-1, nb_classes])

X = tf.placeholder(tf.float32, [None, 16]) #x의 특징이 16개 꼬리가 있다 1, 0 etc.
Y = y_one_hot
#tf.placeholder(tf.float32, [None, nb_classes])

W = tf.Variable(tf.random_normal([16, nb_classes], name = 'weight'))
b = tf.Variable(tf.random_normal([nb_classes]), name = 'bias')

logits = tf.matmul(X, W) + b
hypothesis = tf.nn.softmax(logits)

cost_i = tf.nn.softmax_cross_entropy_with_logits_v2(logits = logits, labels = y_one_hot)

cost = tf.reduce_mean(cost_i)
optimizer = tf.train.GradientDescentOptimizer(learning_rate = 0.1).minimize(cost)

prediction = tf.argmax(hypothesis, 1) #1은 아마 axis
#tf.argmax ; Returns the index with the largest value across axes of a tensor.
#tf.equal ; Returns the truth value of (x == y) element-wise. 그러면 argmax가 int하나가 아니라 배열을 return 하는건가
#참고 axis=0 열 기준, axis=1 행 기준, axis=-1 가장 안쪽에 있는 값
correct_prediction = tf.equal(prediction, tf.argmax(y_one_hot, 1))
accuracy = tf.reduce_mean(tf.cast(correct_prediction, tf.float32))
```

```
with tf.Session() as sess:
  sess.run(tf.global_variables_initializer())
  
  for step in range(2000):
    sess.run(y_one_hot)
    sess.run(optimizer, feed_dict={X: x_data})
    #굳이 바꾸고 싶다면 이렇게 바꾸나
    #강의에서처럼 처리하는 것이 나을 듯
    #강의 Y는 나중에 입력받는 상수, 일단 저장 후 one_hot 처리
    #나는 one_hot 처리 후 저장
    if step % 100 == 0:
      loss, acc = sess.run([cost, accuracy], feed_dict = {X : x_data})
      print("Step : {:5}\tLoss : {:3f}\tAcc : {:.2%}".format(step, loss, acc))
    #예측한 값이 맞는지 아닌지
    #Let's see if we can predict
  pred = sess.run(prediction, feed_dict={X: x_data})
    #y_data: (N, 1) => flatten (N, )
    #[[1], [0]] => [1, 0]
    #zip은 배열 묶듯이 묶는 코드
  for p, y in zip(pred, y_data.flatten()):
    print("[{}] Prediction: {} True Y:{}".format(p == int(y), p, int(y)))
```
